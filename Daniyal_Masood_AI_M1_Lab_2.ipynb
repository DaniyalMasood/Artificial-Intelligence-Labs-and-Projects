```python
# Question 1
#Write a program to find maximum between two numbers.

First_Number = float(input("Please enter first number"))
Second_Number = float(input("Please enter second number"))

if(First_Number > Second_Number):
    
    print("\nThe First Number is Maximum having value : ", First_Number)
    
else:
    
    print("\nThe Second Number is Maximum having value : ", Second_Number)
```

```python
# Question 2
# Write a program to find maximum between three numbers.

First_Number = float(input("Please enter first number : "))
Second_Number = float(input("Please enter second number : "))
Third_Number = float(input("Please enter third number : "))

if(First_Number > Second_Number and First_Number > Third_Number):
    
    print("\nThe First Number is Maximum having value : ", First_Number)
    
elif(Second_Number > First_Number and Second_Number > Third_Number):
    
    print("\nThe Second Number is Maximum having value : ", Second_Number)
    
else:
    
    print("\nThe Third Number is Maximum having value : ", Third_Number)
```

```python
# Question 10

"""
Write a program to input marks (out of 100 each) of five subjects Physics, Chemistry,
Biology, Mathematics and Computer. Calculate percentage and grade of each subject according
to following:
Percentage >= 90% : Grade A
Percentage >= 80% : Grade B
Percentage >= 70% : Grade C
Percentage >= 60% : Grade D
Percentage >= 40% : Grade E
Percentage < 40% : Grade F
"""

Marks_In_Maths = int(input("Enter your Maths Marks : "))
Marks_In_Physics = int(input("Enter your Physics Marks : "))
Marks_In_Chemistry = int(input("Enter your Chemistry Marks : "))
Marks_In_Biology = int(input("Enter your Biology Marks : "))
Marks_In_Computer = int(input("Enter your Computer Marks : "))

Total_Marks = int(input("Enter Total Marks of Matric : "))
Obtain_Marks = (Marks_In_Maths+Marks_In_Physics+Marks_In_Chemistry+Marks_In_Biology+Marks_In_Computer)
Percentage = (Obtain_Marks/Total_Marks)*100

print("\nYour Obtained Marks are : ",Obtain_Marks)
print("Your Percentage is : ",Percentage)

if(Percentage >= 90):
    print("Grade A")
    
elif(Percentage >= 80):
    print("Grade B")
    
elif(Percentage >= 70):
    print("Grade C")
    
elif(Percentage >= 60):
    print("Grade D")
    
elif(Percentage >= 40):
    print("Grade E")
    
else:
    print("Grade F")

```

```python
# Question 11

"""

Write a program to input electricity unit charges and calculate total electricity bill
according to the given condition:
For first 50 units Rs. 0.50/unit
For next 100 units (i.e., 51 to 150) Rs. 0.75/unit
For next 100 units (i.e., 151 to 250) Rs. 1.20/unit
For units above 250 Rs. 1.50/unit
An additional surcharge of 20% is added to the bill

"""

units=int(input("please enter the number of units : "))
if(units<=50):
    Pay_Amount = units*0.5  
    Sub_Charges = Pay_Amount * 0.20
    Total_Bill = Pay_Amount + Sub_Charges
    
elif(units<=51 and units >=150):
    Pay_Amount = units*0.75  
    Sub_Charges = Pay_Amount * 0.20
    Total_Bill = Pay_Amount + Sub_Charges
    
elif(units<=150 and units >=250):
    Pay_Amount = units*1.20  
    Sub_Charges = Pay_Amount * 0.20
    Total_Bill = Pay_Amount + Sub_Charges

elif(units> 250):
    Pay_Amount = units*1.50  
    Sub_Charges = Pay_Amount * 0.20
    Total_Bill = Pay_Amount + Sub_Charges


print("\n Your electricity bill is ", Total_Bill)


```
