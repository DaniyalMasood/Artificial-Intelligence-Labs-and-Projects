```python
# Question 1
# Make a mini-calculator using functions. All the functions should be accessed from module. Your program should ask inputs and option from user. 

choice = 0
print("Please enter (A or a) for addition of two numbers : ")
print("Please enter B for subtraction of two numbers : ")
print("Please enter C for multiplication of two numbers : ")
print("Please enter D for integer division of two numbers : ")
print("Please enter E for float division of two numbers : ")
print("Please enter F for Square of number : ")
print("Please enter G for Square Root of number : ")


while(choice != 'Z'):
            choice = input("\nPlease enter your choice : ")
            if(choice == 'A' or choice == 'a'):
                print("\nYou want Addition of two numbers")
                number1 = int(input("Please enter first number : "))
                number2 = int(input("Please enter second number : "))
                Sum = lab7.Sum(number1, number2)
                print("The Sum of",number1,"and",number2,"is = ",Sum)

            elif(choice == 'B' or choice == 'b'):
                print("\nYou want Subtraction of two numbers")
                number1 = int(input("Please enter first number : "))
                number2 = int(input("Please enter second number : "))
                Difference = lab7.Subtract(number1, number2)
                print("The Subtraction of",number1,"and",number2,"is = ",Difference)

            elif(choice == 'C' or choice == 'c'):
                print("\nYou want Multiplication of two numbers")
                number1 = int(input("Please enter first number : "))
                number2 = int(input("Please enter second number : "))
                Product = lab7.Multiply(number1, number2)
                print("The Multiplication of",number1,"and",number2,"is = ",Product)

            elif(choice == 'D' or choice == 'd'):
                print("\nYou want Integer Division of two numbers")
                number1 = int(input("Please enter first number : "))
                number2 = int(input("Please enter second number : "))
                Int_Division = lab7.IntDivision(number1, number2)
                print("The Integer Division of",number1,"and",number2,"is = ",Int_Division)

            elif(choice == 'E' or choice == 'e'):
                print("\nYou want Float Division of two numbers")
                number1 = int(input("Please enter first number : "))
                number2 = int(input("Please enter second number : "))
                Float_Division = lab7.FloatDivision(number1, number2)
                print("The Float Division of",number1,"and",number2,"is = ",Float_Division)
                
            elif(choice == 'F' or choice == 'f'):
                print("\nYou want Square of numbers")
                number = int(input("Please enter number : "))
                Square = lab7.Square(number)
                print("The Square of",number," is = ",Square)
            
            elif(choice == 'G' or choice == 'g'):
                print("\nYou want Square Root of numbers")
                number = int(input("Please enter number : "))
                Square = lab7.SquareRoot(number)
                print("The Square Root of",number," is = ",Square)
                
            else:
                print("You have exit the program")
                break
```

```python
# Question 2:
"""
Make a module which can return area or volume of desire shape (square, volume
of sphere, area of sphere, area of triangle) Remember area of triangle is not 1‚ÅÑ2(w x h) in
every case
"""

choice = 0
print("Please enter 1 for Calculation of Area of Square : ")
print("Please enter 2 for Calculation of Area of Rectange : ")
print("Please enter 3 for Calculation of Area of Circle : ")
print("Please enter 4 for Calculation of Area of Triangle : ")
print("Please enter 5 for Calculation of Area and Volume of Sphere")
print("Please enter 6 for Calculation of Area and Volume of Cuboid ")

while choice != "Z":
    choice = input("\nPlease enter your choice : ")
    if(choice == "1"):
                print("\nYou want Calculate the Area of Square")
                length = int(input("Please enter length : "))
                Area = lab7.SquareArea(length)
                print("The area of square with side",length,"is = ",Area)

    elif(choice == "2"):
                print("\nYou want Calculate the Area of Rectangle")
                length = int(input("Please enter length : "))
                width = int(input("Please enter width : "))
                Area = lab7.RectangleArea(length, width)
                print("The area of rectangle with length",length," and width",width,"is = ",Area)

    elif(choice == "3"):
                print("\nYou want Calculate the Area of Circle")
                radius = int(input("Please enter radius : "))
                Area = lab7.CircleArea(radius)
                print("The area of circle with radius",radius,"is = ",Area)

    elif(choice == "4"):
                print("\nYou want Calculate the Area of triangle")
                print("Press E for Area of Equilaterial Triangle : ")
                print("Press R for Area of Right Angled and Scalene Triangle : ")
                choice = input("\n Select your choice : ")
                if choice == "R":
                    print("\nYou want Calculate the Area of Right Angled and Scalene Triangle")
                    height = int(input("Please enter height : "))
                    base = int(input("Please enter base : "))
                    Area = lab7.Right_Scalene_TriangeArea(base, height)
                    print("The area of right angle and scalene with base",base," and height",height,"is = ",Area)
                elif choice == "E":
                    print("\nYou want Calculate the Area of Equivalent Triangle")
                    side1 = int(input("Please enter side 1 : "))
                    side2 = int(input("Please enter side 2 : "))
                    side3 = int(input("Please enter side 3 : "))
                    if(side1==side2==side3):
                        Area = lab7.EquilateralTriangleArea(side1, side2, side3)
                        print("The area of equilateral triangle with side1, side2, side3 : ",side1,side2,side3,"is = ",Area)
                    else:
                        print('The triangle is not equilateral triangle')
                    
    elif(choice == "5"):
                print("\nYou want Calculate the Area and Volume of Sphere")
                radius = int(input("Please enter radius : "))
                SurfaceArea = lab7.SphereSurfaceArea(radius)
                Volume = lab7.ShpereVolume(radius)
                print("The surface area of shpere with radius",radius,"is = ",SurfaceArea)
                print("The volume of shpere with radius",radius,"is = ",Volume)
                
    elif(choice == "6"):
                print("\nYou want to calculate the Area and Volume of Cuboid")
                length = int(input("Please enter length : "))
                width = int(input("Please enter width : "))
                height = int(input("Please enter length : "))
                SurfaceArea = lab7.CuboidSurfaceArea(length, width, height)
                Volume = lab7.CuboidVolume(length, width, height)
                print("The surface area of cuboid with length, width and heigth : ",length,width,height,"is = ",SurfaceArea)
                print("The Volume of cuboid with length, width and heigth : ",length,width,height,"is = ",Volume)
                
    else:
                print("You have exit the program")
                break
    
```

```python
# Question 3
# Write a function to print Fibonacci series up to five elements

def PrintFibonacciSeries(number):
    number1, number2, Sum = 0,1,0
    if number < 0:
        print("Your input is wrong")
    else:
        for i in range (0,number):
            Sum = Sum + number1
            print(Sum)
            number1 = number2
            number2 = Sum
    
number = int(input("Please enter the length of series : "))
PrintFibonacciSeries(number)
```
