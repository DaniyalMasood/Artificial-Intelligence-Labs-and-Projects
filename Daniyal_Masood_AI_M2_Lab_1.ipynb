```python
import pandas as pd
import numpy as np
```

```python
"""
Task 1:
Generate two Pandas series. One containing even numbers and other
one consisting of odd numbers.
"""

even = pd.Series(np.arange(0,20,2))
print("Even pandas series : ")
print(even)
print()
odd = pd.Series(np.arange(1,20,2))
print("Odd pandas series : ")
print(odd) 
```

```python
"""
Task 2:
Generate a Data Frame of these two types of numbers (Even and Odd)
and display the table.
"""

even = pd.Series(np.arange(0,20,2))
odd = pd.Series(np.arange(1,20,2))
data={"Even":even , "Odd":odd}
pd_data=pd.DataFrame(data)
print(pd_data) 
```

```python
"""
Task 3:
Create a series of 100 random numbers ranging from 1 to 500 and
display only even values.
"""

Series = pd.Series(np.random.randint(1,500,100))
print(Series)
print("Even Series")
print(Series[Series%2==0])
```

```python
"""
Task 5:
Make a result card of five students in four subjects (Math, Physics,
Chemistry, English) using Data Frame and also add another column of total
marks.
"""
Maths = np.array([100, 98, 96, 94,88])
Physics = np.array([100, 40, 88, 12, 67])
Chemistry = np.array([100, 23, 77, 90, 89])
English = np.array([97, 98, 55, 20, 39])


data = {"Maths":Maths,"Physics":Physics,"Chemistry":Chemistry,"English":English}
df = pd.DataFrame(data, index=['Student1', "Student2","Student3", "Student4", "Student5"])
df['Total'] = df["Maths"]+df["Physics"]+df["Chemistry"]+df["English"]
df["Percentage"] = ((df["Maths"]+df["Physics"]+df["Chemistry"]+df["English"])/400)*100

df
```

```python
import numpy as np
import pandas as pd

X = np.array([78,85,96,80,86])
Y = np.array([84,94,89,83,86])
Z = np.array([86,97,96,72,83])
data = {"X":X,"Y":Y,"Z":Z}
df = pd.DataFrame(data)
df["Total"] = df["X"]+df["Y"]+df["Z"]
df.sort_index(axis = 1)
df.sort_values('X')
df['Percentage change X'] = df['X'].pct_change()
df
```
